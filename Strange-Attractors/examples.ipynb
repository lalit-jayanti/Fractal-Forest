# Strange Attractors and Chaos Games

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from utils.utils import render
from utils.attractors import clifford_attractor, ring_attractor
from utils.chaos_games import jump_game
```

## Examples

```python
img = render(

    num_points = 9000,
    attractor = clifford_attractor(a=1.7, b=1.7, c=0.6, d=1.2),
    image_height = 800,
    image_width = 800,
    sx = 180, sy = 180, rz = np.pi/2,
    initial_iterations = 100,
    max_iterations = 1000
)

plt.figure(figsize=(8,8))
plt.imshow(img, cmap='ocean_r')
plt.axis('off')
plt.savefig('results/clifford_0.png')
```

```python
img = render(

    num_points = 9000,
    attractor = jump_game(n=6, r=2/3, device='cuda') ,
    image_height = 800,
    image_width = 800,
    sx = 400, sy = 400, rz = np.pi/2,
    initial_iterations = 100,
    max_iterations = 1000,
    device = 'cuda'
)

plt.figure(figsize=(8,8))
plt.imshow(img, cmap='ocean_r')
plt.axis('off')
plt.savefig('results/jump_game_6.png')
```
