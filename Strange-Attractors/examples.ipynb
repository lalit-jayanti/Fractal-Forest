# Strange Attractors and Chaos Games

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from utils.utils import render
from utils.attractors import clifford_attractor, ring_attractor
from utils.chaos_games import jump_game
```

## Examples

```python
img = render(

    num_points = 28000,
    attractor = clifford_attractor(a=1.7, b=1.7, c=0.6, d=1.2),
    image_height = 800,
    image_width = 800,
    sx = 180, sy = 180, rz = np.pi/2,
    initial_iterations = 100,
    max_iterations = 2000
)

plt.figure(figsize=(8,8))
plt.imshow(img, cmap='BuGn')
plt.axis('off')
plt.savefig('results/clifford_0.png')
```

```python
img = render(

    num_points = 20000,
    attractor = clifford_attractor(a=2, b=2, c=1, d=-1),
    image_height = 800,
    image_width = 800,
    sx = 180, sy = 180, rz = np.pi/2,
    initial_iterations = 100,
    max_iterations = 2000
)

plt.figure(figsize=(8,8))
plt.imshow(img, cmap='ocean_r')
plt.axis('off')
plt.savefig('results/clifford_1.png')
```

```python
img = render(

    num_points = 15000,
    attractor = ring_attractor(a=1.40, b=1.56, c=1.40 ,d=-6.56),
    image_height = 800,
    image_width = 800,
    sx = 50, sy = 50, rz = np.pi/2,
    initial_iterations = 100,
    max_iterations = 2000
)

plt.figure(figsize=(8,8))
plt.imshow(img, cmap='ocean_r')
plt.axis('off')
plt.savefig('results/ring_attractor_0.png')
```

```python
img = render(

    num_points = 9000,
    attractor = jump_game(n=3, r=1/2, device='cuda') ,
    image_height = 800,
    image_width = 800,
    sx = 400, sy = 400, rz = np.pi/2,
    initial_iterations = 100,
    max_iterations = 1000,
    device = 'cuda'
)

plt.figure(figsize=(8,8))
plt.imshow(img, cmap='pink_r')
plt.axis('off')
plt.savefig('results/jump_game_3.png')
```
