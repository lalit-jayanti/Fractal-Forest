# L systems

```python
import matplotlib.pyplot as plt

from l_systems import LsystemParser
```

## Examples
### G = (V, $\omega$, P)
- V the Alphabet
    - Variables (replacable)
    - Terminals (constants)
- $\omega$ the Axiom or the initiator
- P the set of Production rules
    - predecessor &rarr; successor

```python
# Serpinski triangle
V = {"A":(0.1, 0),"B":(0.1, 0),"+":(0, 60),"-":(0, -60)}
w = "A"
P = {"A":"B-A-B", "B":"A+B+A"}

L = LsystemParser(V, w, P)
w2 = L.generate_string(7)
L.linewidth=1.2
L.visualize(gradient=2)
plt.tight_layout()
plt.savefig("results/serpinski_triangle.png")
```

```python
# Koch curve
V = {"F":(0.1, 0),"+":(0, 60.0),"-":(0, -60.0)}
w = "F--F--F"
P = {"F":"F+F--F+F"}

L = LsystemParser(V, w, P)
w2 = L.generate_string(5)
L.linewidth=1.2
L.visualize(gradient=2)
plt.tight_layout()
plt.savefig("results/koch_curve.png")
```

```python
# Dragon curve
V = {"F":(0.1, 0), "G":(0.1, 0),"+":(0, 90),"-":(0, -90)}
w = "F"
P = {"F":"F+G", "G":"F-G"}

L = LsystemParser(V, w, P)
w2 = L.generate_string(12)
L.linewidth=1.2
L.visualize(gradient=2)
plt.tight_layout()
plt.savefig("results/dragon_curve.png")
```

```python
# Hilbert curve
V = {"A":(0, 0), "B":(0, 0), "F":(1, 0), "+":(0, 90), "-":(0, -90)}
w = "A"
P = {"A":"+BF-AFA-FB+", "B":"-AF+BFB+FA-"}

L = LsystemParser(V, w, P)
w2 = L.generate_string(6)
L.linewidth=2
L.visualize(gradient=2)
plt.tight_layout()
plt.savefig("results/hilbert_curve.png")
```

```python
# Fractal tree
V = {"X":(0, 0), "F":(0.1, 0), "+":(0, 30.0), "-":(0, -45.0)}
w = "X"
P = {"X":"F+[[X]-X]-F[-FX]+X", "F":"FF"}

L = LsystemParser(V, w, P)
w2 = L.generate_string(5)
L.pos[2] = 90.0
L.linewidth=1.2
L.visualize(gradient=2)
plt.tight_layout()
plt.savefig("results/fractal_tree.png")
```

```python
# Binary tree
V = {"A":(0.1, 0), "B":(0.1, 0), "+":(0, 45.0), "-":(0, -90.0)}
w = "A"
P = {"A":"B+[A]-A", "B":"BB"}

L = LsystemParser(V, w, P)
w2 = L.generate_string(7)
L.pos[2] = 90.0
L.linewidth=1.5
L.visualize(gradient=2)
plt.tight_layout()
plt.savefig("results/binary_tree.png")
```

```python
# Gosper curve
V = {"A":(0.1, 0), "B":(0.1, 0), "+":(0, 60), "-":(0, -60)}
w = "A"
P = {"A":"A-B--B+A++AA+B-", "B":"+A-BB--B-A++A+B"}

L = LsystemParser(V, w, P)
w2 = L.generate_string(4)
L.linewidth=1.5
L.visualize(gradient=2)
plt.tight_layout()
plt.savefig("results/gosper_curve.png")
```
